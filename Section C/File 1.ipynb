Q1.3. Create a function that calculates final readiness score from attendance, diary, task, assignment, and lab signals.

---
Ans: Calculate final readiness of students:
1. Define a function named `calculate_readiness_score` taking five inputs: `attendance`, `diary`, `tasks`, `assignment_score` and `lab_score`.
2. Scale each of the inputs and add them together and store the value after rounding it up be two decimals in `total` variable.
3. Apply this function to every row in the data frame and create a new column `readiness_score` which stores the readiness score of each student.



```python
def calculate_readiness_score(attendance, diary, tasks, assignment, lab):

    attendance_score  = (attendance / 100) * 25
    diary_score       = (diary / 15)  * 20
    task_score        = (tasks / 10)  * 20
    assignment_score  = (assignment / 100) * 20
    lab_score         = (lab / 100)  * 15

    total = attendance_score + diary_score + task_score + assignment_score + lab_score
    return round(total, 2)

df['Readiness_Score'] = df.apply(
    lambda row: calculate_readiness_score(
        row['Attendance_Pct'],
        row['Diary_Entries'],
        row['Tasks_Completed'],
        row['Assignment_Score'],
        row['Lab_Score']
    ), axis=1
)

print("Readiness scores calculated:")
df[['Student_Name', 'Readiness_Score']]
```

Q1.4. Use loops or apply logic to classify each student into Support, Developing, Strong, or Excellent.

---
Ans: Classifying each student:

1. Deine a function `classify_student` to categorize students.
2. Using the readiness_score of students, categorize all the students into 'Excellent', 'Strong', 'Developing', 'Support'.
3.  The function creates a new column in the data frame as `performance_band`.



```python
def classify_student(score):
    if score >= 80:
        return 'Excellent'
    elif score >= 65:
        return 'Strong'
    elif score >= 50:
        return 'Developing'
    else:
        return 'Support'

bands = []
for score in df['Readiness_Score']:
    bands.append(classify_student(score))

df['Performance_Band'] = bands

print("Students with performance bands:")
df[['Student_Name', 'Student_Department', 'Readiness_Score', 'Performance_Band']]
```

Q1.5. Generate at least three analytics outputs: department average, top performer, and students needing support.


---
Ans: Analytical Outputs:

1. Calculate and print the average readiness score for each department.
2. Identify and print the details of single top performing student across all departments based on their readiness score.
3. Filter and print the names, departments and readiness scores of all students who have been classified as 'Support' performance band.



```python

print("=== Department Average Readiness Score ===")
dept_avg = df.groupby('Student_Department')['Readiness_Score'].mean().round(2)
print(dept_avg)


print("\n=== Top Performer ===")
top = df.loc[df['Readiness_Score'].idxmax()]
print(f"Name       : {top['Student_Name']}")
print(f"Department : {top['Student_Department']}")
print(f"Score      : {top['Readiness_Score']}")
print(f"Band       : {top['Performance_Band']}")


print("\n=== Students Needing Support ===")
support_students = df[df['Performance_Band'] == 'Support'][
    ['Student_Name', 'Student_Department', 'Readiness_Score']
]
print(support_students.to_string(index=False))
```
